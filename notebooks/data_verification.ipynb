```python
import gzip
from Bio import SeqIO
from Bio import SwissProt
```

```python
with gzip.open(filename='../data/uniprot_sprot.fasta.gz', mode='rt') as handle:
    records = list(SeqIO.parse(handle, 'fasta'))
    print(f"Total counts: {len(records)}")
    
    for i, r in enumerate(records[:3]):
        print(f"\n---Record {i+1}---")
        print(f"ID: {r.id}")
        print(f"Description: {r.description}")
        print(f"Length: {len(r.seq)}")
        print(f"Sequence (first 60 aa): {r.seq[:60]}")
```

```python
with gzip.open("../data/uniprot_sprot.dat.gz", "rt") as handle:
    parser = SwissProt.parse(handle)

    count = 0
    samples = []

    for r in parser:
        count += 1

        if len(samples) < 5:
            samples.append(r)
        else:
            break
    
    for i, r in enumerate(samples):
        print(f"\n---Record {i+1}---")

        acc = r.accessions[0] if r.accessions else None
        print(f"Accession: {acc}")

        print(f"Keywords: {r.keywords}")

        go_terms = [db_id for db, db_id, *rest in r.cross_references if db == "GO"]
        print(f"GO terms: {go_terms}")

        print(f"Organism: {r.organism}")

        print(f"Sequence length: {len(r.sequence)}")

        print(f"Sequence (first 60 aa): {r.sequence[:60]}")
```
